Getting Started with the Dataset

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Directory where the CSV files are located
folder_path = 'C:\\IPL Player Stats\\Batting Stats'

# List to store DataFrames for each year
dfs = []

# Loop through each file in the directory
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Read the CSV file and append it to the list
        file_path = os.path.join(folder_path, file_name)
        df_year = pd.read_csv(file_path)
        dfs.append(df_year)

# Concatenate all DataFrames into a single DataFrame
batting_data = pd.concat(dfs, ignore_index=True)

#Display The First Few rows of the combined dataframe
batting_data.head()
```

```python
# Directory where the CSV files are located
folder_path = 'C:\IPL Player Stats\Bowling Stats'

# List to store DataFrames for each year
dfs = []

# Loop through each file in the directory
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Read the CSV file and append it to the list
        file_path = os.path.join(folder_path, file_name)
        df_year = pd.read_csv(file_path)
        dfs.append(df_year)

# Concatenate all DataFrames into a single DataFrame
bowling_data = pd.concat(dfs, ignore_index=True)

# Display the first few rows of the combined DataFrame
bowling_data.head()
```

```python
# EDA for batting data
batting_data.info()
batting_data.describe()
```

```python
# EDA for bowling data
bowling_data.info()
bowling_data.describe()
```

```python
# Check for missing values
print("Missing values in Batting Data:")
print(batting_data.isnull().sum())
```

```python
print("\nMissing values in Bowling Data:")
print(bowling_data.isnull().sum())
```

Batting Stats Visualization

```python
# Extracting relevant columns for batting stats
top_batsmen = batting_data[['Player', 'Runs']].groupby('Player').sum().sort_values(by='Runs', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_batsmen.index, top_batsmen['Runs'], color='green')
plt.title('Top 10 Batsmen in IPL (2016-2022)')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Bowling Stats Visualization

```python
# Extracting relevant columns for bowling stats
top_bowlers = bowling_data[['Player', 'Wkts']].groupby('Player').sum().sort_values(by='Wkts', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_bowlers.index, top_bowlers['Wkts'], color='blue')
plt.title('Top 10 Bowlers in IPL (2016-2022)')
plt.xlabel('Player')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Bowling Averages Distribution
plt.figure(figsize=(10, 6))
plt.hist(bowling_data['Avg'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Bowling Averages in IPL (2016-2022)')
plt.xlabel('Bowling Average')
plt.ylabel('Frequency')
plt.show()
```

```python
#Economy Rates Distribution
plt.figure(figsize=(10, 6))
plt.hist(bowling_data['Econ'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Economy Rates in IPL (2016-2022)')
plt.xlabel('Economy Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatter Plot - Wickets vs Economy Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wkts', y='Econ', data=bowling_data, color='red')
plt.title('Scatter Plot - Wickets vs Economy Rate in IPL (2016-2022)')
plt.xlabel('Total Wickets')
plt.ylabel('Economy Rate')
plt.show()
```

```python
#Scatter Plot - Wickets vs Strike Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wkts', y='SR', data=bowling_data, color='green')
plt.title('Scatter Plot - Wickets vs Strike Rate in IPL (2016-2022)')
plt.xlabel('Total Wickets')
plt.ylabel('Strike Rate')
plt.show()
```
